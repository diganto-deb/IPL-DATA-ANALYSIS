# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# DATA READING

```python
df=pd.read_csv('Dataset/matches.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
#Check for null values
sb.heatmap(df.isnull(),cmap='viridis')
```

```python
#check for null values of 'winner' coulmn
df[pd.isnull(df['winner'])]
```

```python
#Fill the null values of 'winner' column with value 'draw'
df['winner'].fillna('Draw', inplace=True)
```

```python
df.head()
```

```python
#Find the citiws which are null
df[pd.isnull(df['city'])]
```

```python
df['city'].fillna('Dubai',inplace=True)
```

```python
sb.heatmap(df.isnull(),cmap='viridis')
```

# 1.Which stadium is best for winning by runs?

```python
df[df['win_by_runs']!=0]['venue'].mode()
```

# 2. Which stadium is best for winning by wickets?

```python
df[df['win_by_wickets']!=0]['venue'].mode()
```

# 3. For a IPL team(take any team of your choice) which stadium is best when they win the toss?

```python
df[(df['toss_winner']=='Sunrisers Hyderabad') & (df['winner']=='Sunrisers Hyderabad')]['venue'].mode()
```

# 4.Which team is best for chasing?

```python
df[df['win_by_wickets']!=0]['winner'].mode()
```

# 5.Which team take fielding and won most of the matches?

```python
df[df['toss_decision']=='field']['winner'].mode()
```

# 6.Which is the best defending team

```python
df[df['win_by_runs']!=0]['winner'].mode()
```

# 7.Has Toss-winning help match winning?

```python
plt.figure(figsize=(12,7))
to=df['toss_winner']==df['winner']
sb.countplot(to)
```

From the above countplot it looks like, Toss winning actually helps in Match winning – or to be statistically right, we could say there’s a correlation between Toss Winning and Match Winning and so we can assume that it helps.

# 8.Which team won more when won toss?

```python
plt.figure(figsize=(30,10))
sb.countplot(df.winner[df.toss_winner==df.winner])
plt.show()
```

From the above graph , We can see that Mumbai Indians won more mmatches when won toss

# 14.Which is best chasing venue?

```python
df[df['win_by_wickets']!=0]['venue'].mode()
```

```python
plt.figure(figsize=(5,12))
sb.swarmplot(x='win_by_wickets',y='venue',data=df)
plt.show()
```

# Eden Gardens and M Chinnaswamy Stadium are the best venues for chasing the density of each attribute contributes in swarmplot figure.

# Heatmap

```python
cor=df.corr()
plt.figure(figsize=(12,10))
sb.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```

# From the above heatmap diagram there is a correlation between the win_by_runs and win_by_wickets ,and we know that if one the value is zero then other would be non-zero value.

# Most successful ipl team ?

```python
plt.figure(figsize=(50,60))
sb.factorplot(y='winner',kind='count',data=df)
plt.show()
```

# From the above factorplot data we can say that most winning team is the most successful team and Mumbai Indians is the most successful team followed by Chennai super kings and Kolkata knight riders.

# Team won by max runs and best defending team?

```python
plt.figure(figsize=(10,7))
sb.swarmplot(x='win_by_runs',y='winner',data=df)
plt.show()
```

# From the above swarmplot data we can say that Mumbai Indians won by maximum runs in ipl history.And from the above data ,best defending team is Mumbai Indians.


# Team won by max wickets?and best chasing team? 


```python
plt.figure(figsize=(5,12))
sb.swarmplot(x='win_by_wickets',y='winner',data=df)
plt.show()
```

# From the above data we can say Kolkata knight riders are best chasing team and most of the teams won by 9 wickets but royal challengers bangalore won two times by maximum wickets taken.
